# kmp  
  

**1nextnext[j]1-j**  
```c
#include <iostream>
#include <cstring>
using namespace std;
const int N = 100010, M = 1000010;
char p[N], s[M];
int n, m, ne[N];

int main()
{
    cin >> n >> p + 1 >> m >> s + 1;
    memset(ne, 0, sizeof(ne));
    //next next[1] = 0
    for(int i = 2, j = 0;i <= n;i++)
    {
        //jj!=0j
        while(j && p[j + 1] != p[i]) j = ne[j];
        //ij
        if(p[j+1] == p[i]) j++;
        ne[i] = j;
    }

    //kmp
    for(int i = 1, j = 0;i <= m;i++)
    {
        while(j && p[j + 1] != s[i]) j = ne[j];
        if(p[j + 1] == s[i]) j++;
        if(j == n)
        {
            //
            cout << i - j << " ";
            //
            j = ne[j];
        }
    }
    return 0;
}
```

## Trie  


```c
#include <iostream>

using namespace std;
const int N = 100010;

int tr[N][26], cnt[N];
int idx = 0;

void insert(char* str)
{
    //
    int p = 0;
    for(int i = 0;str[i];i++)
    {
        int k = str[i] - 'a';
        if(!tr[p][k]) tr[p][k] = ++idx;
        p = tr[p][k];
    }
    cnt[p]++;
}

int query(char* str)
{
    int p = 0;
    for(int i = 0;str[i];i++)
    {
        int k = str[i] - 'a';
        if(!tr[p][k]) return 0;
        p = tr[p][k];
    }
    return cnt[p];
}

int main()
{
    char op[2], str[N];
    int n;
    cin >> n;
    while(n--)
    {
        cin >> op;
        if(op[0] == 'I')
        {
            cin >> str;
            insert(str);
        }else
        {
            cin >> str;
            int res = query(str);
            cout << res << endl;
        }
    }
    return 0;
}
```

```c
#include <iostream>
typedef unsigned long long ULL;
using namespace std;
const int N = 100010;
//hp
int p[N], h[N], k = 131;
char str[N];

ULL get(int l, int r)
{
    //1234h[2]12h[4]123434h[4] - h[2]*100
    return h[r] - h[l -  1]*p[r - l  + 1];
}

int main()
{
    int n, m;
    cin >> n >> m;
    scanf("%s", str + 1);
    p[0] = 1;
    for(int i = 1;i <= n;i++)
    {
        p[i] = p[i - 1]*k;
        h[i] = h[i - 1]*k + str[i];
    }
    while(m--)
    {
        int l1, r1, l2, r2;
        cin >> l1 >> r1 >> l2 >> r2;
        if(get(l1, r1) == get(l2, r2))
        puts("Yes");
        else
        puts("No");
    }
    return 0;
}
```
